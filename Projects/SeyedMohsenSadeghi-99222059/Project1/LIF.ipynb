```python
import math
from LIF import LIF
```

```python
def i_interval(time, i=5, duration=(20, 80)):
    """I(t) = c"""
    start, end = duration
    if start < time < end:
        return i
    return 0


def i_linear(time, c=2):
    """I(t) = c.t"""
    return c * time


def i_sine(time, c=10):
    """I(t) = c.(sin(t) + 0.9)"""
    return c * (0.9 + math.sin(time))


def i_constant(time, i=5):
    """I(t) = c"""
    return i


def i_step(time, i=5, d1=(20, 30), d2=(40, 70)):
    """I(t) = c"""
    s1, e1 = d1
    s2, e2 = d2
    if s1 < time < e1 or s2 < time < e2:
        return i
    return 0
```

```python
model1 = LIF(input_current=i_sine, is_adaptive=False, is_exponential=False)
model1.u_t_plot()
model1.w_t_plot()
model1.f_i_plot()
```

```python
model2 = LIF(input_current=i_sine, is_adaptive=True, is_exponential=False)
model2.u_t_plot()
model2.w_t_plot()
model2.f_i_plot()
```
