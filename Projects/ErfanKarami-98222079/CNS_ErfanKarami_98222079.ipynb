```python
from PyNeuron import LIF,ALIF,AELIF,AdaptiveMCH
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
def const_current(time):
    return 4000
```

```python
def sine_current(time):
    return 4000*(math.sin(time)+0.9)
```

```python
def random_current(time):
    if np.random.rand()>0.7:
        return 25000*np.random.rand()
    else:
        return 0
```

```python
def step_current(time):
    if np.random.rand()>0.95:
        return 50000
    else: return 0
```

```python
def linear_current(time):
    return 100*time+3500
```

```python
lif=LIF(10,8,-79,-50,5,-68)
```

```python
lif_const=lif.process(const_current,100,0.03125)
```

```python
voltage_data=lif_const['voltage']
spikes=lif_const['spikes']
```

```python
print('spikes:\n',spikes)
```

```python
# LIF with const current
figure=plt.figure(figsize=(10,15))

figure.suptitle('LIF with const current',fontsize=16)

ax1=figure.add_subplot(311)
ax1.plot(voltage_data[:,0],voltage_data[:,1])
ax1.set_ylabel("membrane potential (mV)")
ax1.set_xlabel("time (mS)")

I=np.zeros_like(voltage_data[:,0])
for i in range(len(I)):
    I[i]=const_current(voltage_data[i,0])


ax2=figure.add_subplot(312)
ax2.plot(voltage_data[:,0],I)
ax2.set_ylabel("input current(pA)")
ax2.set_xlabel("time (mS)")

plt.plot(voltage_data[:,0],I)

ax3=figure.add_subplot(313)
ax3.text(0,0.85,'total time frame: 100 mS',fontsize=13)
ax3.text(0,0.75,'dt: 0.03125 mS',fontsize=13)
ax3.text(0,0.65,'initial refactory time: 0 mS',fontsize=13)
ax3.text(0,0.55,'refactory period: 0 mS',fontsize=13)
ax3.text(0.5,0.85,'R: 10 M ohm',fontsize=13)
ax3.text(0.5,0.75,'tau: 8 mS',fontsize=13)
ax3.text(0.5,0.65,'u_rest: -79 mV',fontsize=13)
ax3.text(0.5,0.55,'threshold: -50 mV',fontsize=13)
ax3.text(0.5,0.45,'u_spike: 5 mV',fontsize=13)
ax3.text(0.5,0.35,'u_reset: -68 mV',fontsize=13)
ax3.set_axis_off()

plt.show()
```
